```python
import matplotlib.pyplot as plt
import seaborn as sns

# Light clean theme
sns.set_theme(style="whitegrid")  
plt.rcParams["axes.facecolor"] = "#E9F4FF"      # very light blue
plt.rcParams["figure.facecolor"] = "white"

# Strong pastel colors (more visible than default pastel)
strong_pastel = ["#4DB6E8", "#FF9AA2", "#FFC4A3", "#B5EAD7", "#C7CEEA", "#FFDAC1"]

# Column groups
ordered_cat = ['day_of_week']
nominal_cat = ['network_type', 'bank_status', 'device_type', 'Payment_Application', 'Status']
binary_cat = ['upi_switch_status', 'app_available']
numerical = ['internet_strength', 'server_load_percent', 'transaction_amount', 'internet_latency']

plt.figure(figsize=(16, 20))
plot_num = 1

plt.suptitle("Online Payment App Insights", fontsize=18, y=1.02)

# Ordered categorical (line)
for col in ordered_cat:
    plt.subplot(6, 2, plot_num)
    order = ['Monday', 'Tuesday', 'Wednesday', 'Thursday', 'Friday', 'Saturday', 'Sunday']
    counts = data[col].value_counts().reindex(order)
    plt.plot(counts.index, counts.values, marker='o', linestyle='-', color="#1f77b4")
    plt.title(f'Line Plot of {col}')
    plot_num += 1

# Nominal categorical (MORE VISIBLE BAR COLORS)
for col in nominal_cat:
    plt.subplot(6, 2, plot_num)
    sns.countplot(
        y=col,
        data=data,
        order=data[col].value_counts().index,
        palette=strong_pastel   # ⬅️ Stronger colors
    )
    plt.title(f'Count Plot of {col}')
    plot_num += 1

# Binary categorical (pie)
for col in binary_cat:
    plt.subplot(6, 2, plot_num)
    data[col].value_counts().plot.pie(
        autopct='%1.1f%%',
        colors=['#66b3ff', '#ff9999'],
        textprops={'color': 'black'}
    )
    plt.title(f'Pie Chart of {col}')
    plt.ylabel('')
    plot_num += 1

# Numerical (histogram)
for col in numerical:
    plt.subplot(6, 2, plot_num)
    sns.histplot(data[col], kde=True, color="lightblue")
    plt.title(f'Histogram + KDE of {col}')
    plot_num += 1

plt.tight_layout()
plt.show()
```

```python
# Encoder objects


from sklearn.preprocessing import LabelEncoder

le_day = LabelEncoder()
le_network = LabelEncoder()
le_strength = LabelEncoder()
le_bank_status = LabelEncoder()
le_device = LabelEncoder()
le_upi = LabelEncoder()
le_app = LabelEncoder()
le_payment_app = LabelEncoder()
le_status = LabelEncoder()

# Applying encoders one-by-one
data['day_of_week'] = le_day.fit_transform(data['day_of_week'])
data['network_type'] = le_network.fit_transform(data['network_type'])
data['internet_strength'] = le_strength.fit_transform(data['internet_strength'])
data['bank_status'] = le_bank_status.fit_transform(data['bank_status'])
data['device_type'] = le_device.fit_transform(data['device_type'])
data['upi_switch_status'] = le_upi.fit_transform(data['upi_switch_status'])
data['app_available'] = le_app.fit_transform(data['app_available'])
data['Payment_Application'] = le_payment_app.fit_transform(data['Payment_Application'])
data['Status'] = le_status.fit_transform(data['Status'])
```

```python
from plotly.subplots import make_subplots
from plotly import graph_objects as go
import plotly.io as pio   

def box_plot(columns):
    # Create subplots
    fig = make_subplots(rows=1, cols=len(columns))

    # Iterate over columns
    for i, col in enumerate(columns):

        # Create Box trace
        box_trace = go.Box(y=data[col], name=col )

        # Add Box trace to subplot
        fig.add_trace(box_trace, row=1, col=i+1)

        fig.update_xaxes(title_text=col, row=1, col=i+1)

    # Update subplot layout
    fig.update_layout(title='Box Plots', height=600,width=250*len(columns))
    fig.update_yaxes(title_text='Values')
    fig.write_image("boxplots.png") 
    return fig.show()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,random_state=42,test_size=0.2) 
```

```python
print(X_train.shape,"X_train_values")
print(X_test.shape,"X_test_values")
print(y_train.shape,"y_train_values")
print(y_test.shape,"y_test_values")
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter

smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_train, y_train)

print("After SMOTE:", Counter(y_resampled))
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler=StandardScaler()
X_train_scaled=scaler.fit_transform(X_resampled)
X_test_scaled=scaler.transform(X_test)
```

```python
#Importing necessary libraries
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC

from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier

from sklearn.metrics import confusion_matrix,classification_report
```

```python
#initialising the classifiers
lr = LogisticRegression()
GNB = GaussianNB()
DTC = DecisionTreeClassifier(random_state=42)
KNN = KNeighborsClassifier(n_neighbors=15)
SVM = SVC(random_state=42)

RFC=RandomForestClassifier()
Adaboost=AdaBoostClassifier()
GBC=GradientBoostingClassifier()
```

```python
#Applying Logistical Regression
lr.fit(X_resampled, y_resampled)
y_pred=lr.predict(X_test)
print("Measuring Efficiency of Logistic Regression\n")
print("Confusion Matrix is :\n",confusion_matrix(y_test,y_pred))
print('classification report is:\n',classification_report(y_test,y_pred))
```

```python
#Applying gaussian naivebayes
GNB.fit(X_resampled, y_resampled)
y_pred=GNB.predict(X_test)
print("Measuring Efficiency of GaussianNB\n")
print("Confusion Matrix is :\n",confusion_matrix(y_test,y_pred))
print('classification report is:\n',classification_report(y_test,y_pred))
```

```python
#Applying Decisiontree classifier
DTC.fit(X_resampled, y_resampled)
y_pred=DTC.predict(X_test)
print("Measuring Efficiency of Decisiontree classifier\n")
print("Confusion Matrix is :\n",confusion_matrix(y_test,y_pred))
print('classification report is:\n',classification_report(y_test,y_pred))
```

```python
# Applying K nearest nieghbors
KNN.fit(X_resampled, y_resampled)
y_pred=KNN.predict(X_test)
print("Measuring Efficiency of KNN\n")
print("Confusion Matrix is :\n",confusion_matrix(y_test,y_pred))
print('classification report is:\n',classification_report(y_test,y_pred))
```

```python
#Applying Support Vector Machine Classifier
SVM.fit(X_resampled, y_resampled)
y_pred=SVM.predict(X_test)
print("Measuring Efficiency of Support Vector Machine\n")
print("Confusion Matrix is :\n",confusion_matrix(y_test,y_pred))
print('classification report is:\n',classification_report(y_test,y_pred))
```

```python
#Applying Randomforest classifier
RFC.fit(X_resampled, y_resampled)
y_pred=RFC.predict(X_test)
print("Measuring Efficiency of Randomforest classifier\n")
print("Confusion Matrix is :\n",confusion_matrix(y_test,y_pred))
print('classification report is:\n',classification_report(y_test,y_pred))
```

```python
#Applying  Adaboost classifier
Adaboost.fit(X_resampled, y_resampled)
y_pred=Adaboost.predict(X_test)
print("Measuring Efficiency of Adaboost classifier\n")
print("Confusion Matrix is :\n",confusion_matrix(y_test,y_pred))
print('classification report is:\n',classification_report(y_test,y_pred))
```

```python
#Applying Gradientboosting classifier
GBC.fit(X_resampled, y_resampled)
y_pred=GBC.predict(X_test)
print("Measuring Efficiency of  Gradientboosting classifier\n")
print("Confusion Matrix is :\n",confusion_matrix(y_test,y_pred))
print('classification report is:\n',classification_report(y_test,y_pred))
```

```python
from sklearn.metrics import accuracy_score, precision_score
```

```python
models = [lr, SVM, KNN, DTC, RFC, GBC, GNB, Adaboost]

model_names = ["LR", "SVM", "KNN", "DTC", "RFC", "GBC", "GNB", "Adaboost"]

acc_scores = []
pre_scores = []

for model in models:
    y_pred = model.predict(X_test)
    acc_scores.append(accuracy_score(y_test, y_pred))
    pre_scores.append(precision_score(y_test, y_pred))

final_data = pd.DataFrame({
    "Models": model_names,
    "ACC": acc_scores,
    "PRE": pre_scores
})
```

```python
final_data
```

```python
max_acc = final_data["ACC"].max()

# Highlight models within 0.0006 of max accuracy
colors = ["orange" if acc >= max_acc - 0.0006 else "#4DB6E8" 
          for acc in final_data["ACC"]]

plt.figure(figsize=(8, 4))
sns.barplot(data=final_data, x="Models", y="ACC", palette=colors)

plt.ylim(0, 1)
plt.xticks(fontsize=12, fontweight="bold")
plt.yticks(fontsize=12, fontweight="bold")
plt.title("Model Accuracy Comparison", fontsize=14, fontweight="bold")
plt.gca().grid(False)   # remove gridlines

plt.show()
```
